```python
def box_age(cols):
    age=cols[0]
    pclass=cols[1]
    if pd.isnull(age):
        if pclass==1:
            return 37
        elif pclass==2:
            return 29
        else:
            return 24
    else:
        return age
            
    
```

```python
tdf['Age']=tdf[['Age','Pclass']].apply(box_age,axis=1)
sns.heatmap(tdf.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
tdf
```

```python
tdf.drop('Cabin',axis=1,inplace=True)
```

```python
tdf
```

```python
tdf.dropna(inplace=True)
```

```python
tdf['Embarked'].isnull()
```

```python
pd.get_dummies(tdf['Sex']) #To catogrize the given data in 0s & 1s format (Digital Format)
```

```python
#taking male data only
sm=pd.get_dummies(tdf['Sex'],drop_first=True)
sm
```

```python
pd.get_dummies(tdf['Embarked'])
```

```python
em=pd.get_dummies(tdf['Embarked'],drop_first=True)
em
```

```python
td=pd.concat([tdf,sm,em],axis=1)
td
```

```python
td.head(2)
```

```python
td.drop(['PassengerId','Name','Sex','Embarked','Ticket'], inplace=True, axis=1)
td
```

```python
x=td.drop('Survived', axis=1)
y=td['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Random_state:- Accuraccy check, if not data will be hard to get

x_tr,x_ts,y_tr,y_ts=train_test_split(x,y,test_size=0.2,random_state=101) 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(solver='lbfgs', max_iter=200)
```

```python
lr.fit(x_tr,y_tr)
```

```python
predict=lr.predict(x_ts)
```

```python
from sklearn.metrics import classification_report as cr
```

```python
print(cr(y_ts,predict)) #0 is false 1 is true
```

```python
from sklearn.metrics import confusion_matrix as cf
```

```python
cf(y_ts,predict) #178
```
